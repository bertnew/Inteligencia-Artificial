```python
# 1. Se importa 'transformers' modelo de deep learning diseñado
# principalmente para manejar datos secuenciales

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# clase: AutoTokenizer: sirve para cargar automaticamente el tokenizador
# adecuado para una tarea especifica sin necesidad de especificar una ruta
# exacta del modelo.

# clase: AutoModelForSeq2SeqLM: una clase de modelo de la libreria transformer
```

```python
# Helsinki-NLP/opus-mt: un modelo entrenado de traduccion preentrenado

# Cargar el tokenizador y el modelo para la traducción de español a inglés

tokenizer_es_en = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-es-en")
model_es_en = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-es-en")

# Cargar el tokenizador y el modelo del chatbot
tokenizer_chatbot = AutoTokenizer.from_pretrained("facebook/blenderbot_small-90M")
model_chatbot = AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot_small-90M")
```

```python
while True:
    # Pedir entrada al usuario en español
    user_input_es = input("\nPaciente/Tú: ")

    # Salir del bucle si el usuario decide terminar
    if user_input_es.lower() in ["exit", "salir", "terminar", 'adiós', 'cerrar', 'end']:
        print("¡Hasta luego!")
        break   # instruccion que sirve para salir, se detiene la ejecucion cuando el usuario escdriba 'exit' 'adios' etc y finaliza la conversacion

    # Traducir la entrada del usuario de español a inglés
    input_ids_es_en = tokenizer_es_en(user_input_es, return_tensors="pt").input_ids # es la entrada del texto que la salida de la funcin debe ser de tensores PyTorch que facilita su uso posterior del modelo
    output_es_en = model_es_en.generate(input_ids_es_en)
    translated_input = tokenizer_es_en.decode(output_es_en[0], skip_special_tokens=True)



#  [0]: Esto indica que estamos accediendo al primer resultado generado por el modelo. En este caso, suponemos que solo se genera una salida.
# Cuando skip_special_tokens se establece en True, estos tokens especiales se omiten del texto decodificado, lo que resulta en un texto más limpio y legible.

   # Tokenizar la entrada traducida
    input_ids = tokenizer_chatbot(translated_input, return_tensors="pt").input_ids
# Esta línea de código utiliza el tokenizador del chatbot para convertir la entrada traducida en una secuencia de tokens y luego devuelve esos tokens como tensores de PyTorch.
# El parámetro return_tensors="pt" se utiliza al tokenizar secuencias de texto con el tokenizador de Hugging Face
# Cuando return_tensors="pt", el tokenizador devuelve los tokens como tensores de PyTorch. Esto es útil cuando se trabaja con modelos de PyTorch, ya que permite alimentar directamente los tokens tokenizados al modelo para su procesamiento.

    # Generar la salida del modelo del chatbot
    output = model_chatbot.generate(input_ids)

# output: Es una variable que almacenará la salida generada por el modelo.
# model_chatbot: Es el modelo utilizado para generar la salida del chatbot.
    # Decodificar y mostrar la respuesta del modelo del chatbot
    decoded_output = tokenizer_chatbot.decode(output[0], skip_special_tokens=True)
    print(f"Psicologo(a): {decoded_output}")


# decode output: La línea de código que proporcionaste decodifica la salida generada por el modelo de chatbot utilizando el tokenizador.
# decoded_output: Es una variable que almacenará la salida decodificada del modelo.
# tokenizer_chatbot: Es el tokenizador utilizado para tokenizar las entradas y decodificar las salidas del modelo.
# decode(output[0], skip_special_tokens=True): Esta es la función que decodifica
# los tokens de la salida generada por el modelo.
# output[0] representa la primera salida del modelo, y skip_special_tokens=True indica que los tokens especiales (como los tokens de inicio y fin de secuencia) deben ser omitidos en la salida decodificada. Esto es útil para obtener la respuesta del modelo de una manera más limpia y legible.
```
